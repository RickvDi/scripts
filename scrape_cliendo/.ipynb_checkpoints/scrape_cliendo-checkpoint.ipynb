```python
xpaths = {
    "dagrapportage": "/html/body/div[1]/div[2]/div/div/div[1]/div[2]/div[4]/div/ul/li[1]/a",
    
    "intake_btn": "/html/body/div[1]/div[2]/div/div/div[2]/div/div[1]/div[2]/div[1]/div[2]/div/div[1]/div[1]",
    "pers_btn": "/html/body/div[1]/div[2]/div/div/div[2]/div/div[1]/div[2]/div[1]/div[2]/div/div[1]/div[2]",
    "doelen_btn": "/html/body/div[1]/div[2]/div/div/div[2]/div/div[1]/div[2]/div[1]/div[2]/div/div[1]/div[3]",
    "persoonsbeeld_btn": "/html/body/div[1]/div[2]/div/div/div[2]/div/div[1]/div[2]/div[1]/div[2]/div/div[1]/div[4]",
    
    "belangrijke_info": "/html/body/div[1]/div[2]/div/div/div[2]/div/div[1]/div[1]/div[2]/div[1]/div[4]/div[1]/div/div[2]/textarea",
    "organisatie": "/html/body/div[1]/div[2]/div/div/div[1]/div[2]/div[1]/div[2]/div[3]/a",
    'intake_datum': "/html/body/div[1]/div[2]/div/div/div[2]/div/div[1]/div[1]/div[2]/div[1]/div[4]/div[2]/div/div[2]/input",
    
    "vragen_formulier": "/html/body/div[1]/div[2]/div/div/div[2]/div/div[1]/div[2]/div[2]/div[1]/div[2]",
    
}


def move_to_page(persoon_url, extensie=None):
    if extensie is None:
        driver.get(f"https://secure.cliendo.com/i801/{persoon_url}")
    else:
        driver.get(f"https://secure.cliendo.com/i801/{persoon_url}/{extensie}")
    driver.implicitly_wait(3)
    
def click_button(btn):
    driver.find_element(By.XPATH, xpaths[btn]).click()

def get_data(item):
    return driver.find_element(By.XPATH, xpaths[item]).get_attribute('value')

def get_text(item):
    return driver.find_element(By.XPATH, xpaths[item]).get_attribute('text')

def get_children_class(class_name):
    return driver.find_elements(By.CLASS_NAME, class_name)

def get_children_xpath(elem):
    return driver.find_elements(By.XPATH, xpaths[elem])


# ga naar formulier
move_to_page('rico-lammers', 'details')    
click_button("intake_btn")
formulier = driver.find_element(By.XPATH, xpaths['vragen_formulier'])
```

```python
def get_xpath(element) -> str:
    """
    Genereer het absolute XPath voor een gegeven WebElement.
    """
    components = []
    child = element
    while child is not None:
        parent = child.find_element(By.XPATH, '..')
        children = parent.find_elements(By.XPATH, '*')
        index = 1
        for i in range(len(children)):
            if children[i] == child:
                components.append(child.tag_name + f'[{index}]')
                break
            if children[i].tag_name == child.tag_name:
                index += 1
        child = parent
        if child.tag_name == 'html':
            break
    components.reverse()
    return '/' + '/'.join(components)
```

```python
move_to_page('rico-lammers', 'details')   
#move_to_page('elise-peeters', 'details')   
click_button("intake_btn")
formulier = driver.find_element(By.XPATH, xpaths['vragen_formulier'])

# ga alle categorien langs
antwoord_text = {"a": "Nooit",
                "b": "Bijna nooit",
                "c": "Soms",
                "d": "Bijna altijd",
                "e": "Altijd",
                "f": "Weet ik niet",
                "7": "Ligt aan de situatie, namelijk"}

n_vragen = {0: 6,
           1: 11,
           2: 5,
           3: 3,
           4: 6,
           5: 3}
vragen_namen = [(it, f"Emz-AvoUit-1{cat}{vraag}") for it, cat in enumerate('abcdef') for vraag in "abcdefghijklmnopqr"]

json_vragen = []

cat_nr = 0

vragenlijst_huidig = {}
is_filled = True
for cat in formulier.find_elements(By.CLASS_NAME, 'subGroup'):
    title = cat.find_element(By.CLASS_NAME, "title").find_element(By.XPATH, ".//h3").text
    if "In hoeverre herken" in title or "Tags" in title:
        continue
    
    vragen_cat = [vraag for it, vraag in vragen_namen if it == cat_nr]
    json_cat = []
    for vraag_nr, vraag in enumerate(vragen_cat):
        # niet elke categorie heeft evenveel vragen, skip vragen die hierboven wel hard coded gedefieerd zijn
        if vraag_nr > n_vragen[cat_nr]:
            break
        try:
            # Zoek alle radio buttons met dezelfde naam
            radio_buttons = cat.find_elements(By.NAME, vraag)
            if len(radio_buttons) == 0:
                continue
            # Controleer welke radio button geselecteerd is
            selected_value = None
            for radio_button in radio_buttons:
                if radio_button.is_selected():
                    selected_value = radio_button.get_attribute('value')
                    break
            if selected_value is None:
                vragenlijst_huidig = None
                break
            vraag = radio_button.find_element(By.XPATH, '../../../..').find_element(By.XPATH, './/div[1]/span[1]').text
            # Print de geselecteerde waarde
            if selected_value == "7":
                # meestal staat text op een div[13] maar sporadisch op div[15] dus er moet bepaald worden welke div index erbij hoort
                xpath = get_xpath(radio_button).split("/")[-5]
                # haal cijfer uit div[cijfer]
                cijfer = int(xpath[4:-1])
                parent = radio_button.find_element(By.XPATH, '../../../../..')
                txt = parent.find_element(By.XPATH, f'.//div[{cijfer+1}]/div[1]/textarea').get_attribute('value')
                json_vraag = {f"{vraag_nr}": {"vraag": vraag, "antwoord": antwoord_text[selected_value[-1]], "toelichting": txt}}
            else:
                json_vraag = {f"{vraag_nr}": {"vraag": vraag, "antwoord": antwoord_text[selected_value[-1]]}}

            json_cat.append(json_vraag)
        except:
            pass
    
    try:
        eigen_text = formulier.find_element(By.CLASS_NAME, vragen_cat[vraag_nr + 1]).find_element(By.XPATH, './/div[1]/textarea')
        overige_text = formulier.find_element(By.CLASS_NAME, vragen_cat[vraag_nr + 3]).find_element(By.XPATH, './/div[1]/textarea')
        eigen_input = eigen_text.get_attribute('value')
        overige_text = overige_text.get_attribute('value')
        vragenlijst_huidig[f"{cat_nr}"] = {"titel": title, "vragen": json_cat, 'input_deelnemer': eigen_input, 'input_begeleiding': overige_text}
    except:
        pass
    cat_nr += 1 
    

print("done")
```
